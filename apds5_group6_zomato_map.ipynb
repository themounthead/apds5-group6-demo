```python
df_folium = pd.DataFrame({'Lat':zomato_cuisine_data['latitude'],'Long':zomato_cuisine_data['longitude']})

# df_folium['weight'] = df_folium['Count'] / df_folium['Count'].abs().max()

def generateBaseMap(loc, zoom=11, tiles='OpenStreetMap', crs='ESPG2263'):
    return folium.Map(location=loc,
                   control_scale=True, 
                   zoom_start=zoom,
                   tiles=tiles)
  
base_map = generateBaseMap([28.5, 77.25] )

map_values1 = df_folium[['Lat','Long']]

data = map_values1.values.tolist()
           
hm = HeatMap(data,gradient={0.1: 'blue', 0.3: 'green', 0.5: 'yellow', 0.7: 'orange', 1: 'red'}, 
                min_opacity=0.05, 
                max_opacity=0.9, 
                radius=50,
                use_local_extrema=False)
#.add_to(base_map)

base_map.add_child(hm)
```
